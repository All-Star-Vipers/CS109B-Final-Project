```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import math
```

```python
df_tmdb = pd.read_csv('../Milestone 1/detailed_movie_data_tmdb.csv')
```

```python
df_genres = pd.read_csv('../Milestone 1/tmdb_genre_encoding.csv')
```

```python
df_tmdb = df_tmdb[df_tmdb['original_language'] == 'en']
df_tmdb = df_tmdb[df_tmdb['status'] == 'Released']
df_tmdb = df_tmdb[~df_tmdb['poster_path'].isnull()]

columns_to_list = ['tagline', 'title']
for column in columns_to_list:
    df_tmdb[column] = df_tmdb[column].fillna('')
    df_tmdb[column] = df_tmdb[column].apply(lambda x: x if '[' in x else [x])
```

```python
df_tmdb.columns
```

```python
df_genres = df_genres.merge(df_tmdb[['id', 'belongs_to_collection']], on = 'id')
```

```python
no_collection = df_genres[df_genres['belongs_to_collection'].isnull()].sum().drop(['id','belongs_to_collection'])
collection = df_genres[~df_genres['belongs_to_collection'].isnull()].sum().drop(['id','belongs_to_collection'])
collection_pct = [series * 1.0 / sum(collection) for series in collection]
no_collection_pct = [series * 1.0 / sum(no_collection) for series in no_collection]
collection_diff = [i - j for i,j in zip(collection_pct, no_collection_pct)]
plt.bar(range(len(collection)), collection_diff)
plt.xlabel(collection.keys())
```

```python
df_tmdb['part_of_collection'] = df_tmdb['belongs_to_collection'].notnull()
```

```python
df_tmdb['title_match'] = (df_tmdb['original_title'] != df_tmdb['title'])
```

```python
df_tmdb['title_match'].value_counts()
```

```python
df_genres = df_genres.merge(df_tmdb[['id', 'title_match']], on = 'id')
```

```python
no_collection = df_genres[df_genres['title_match'] == True].sum().drop(['id','title_match'])
collection = df_genres[df_genres['title_match'] == False].sum().drop(['id','title_match'])
collection_pct = [series * 1.0 / sum(collection) for series in collection]
no_collection_pct = [series * 1.0 / sum(no_collection) for series in no_collection]
collection_diff = [i - j for i,j in zip(collection_pct, no_collection_pct)]
plt.bar(range(len(collection)), collection_diff)
plt.xlabel(collection.keys())
```
